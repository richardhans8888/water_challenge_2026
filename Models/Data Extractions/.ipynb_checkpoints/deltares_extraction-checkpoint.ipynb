```python
import requests
import planetary_computer as pc
import xarray as xr

# URL for the Deltares Water Availability Dataset
url = "https://planetarycomputer.microsoft.com/api/stac/v1/collections/deltares-water-availability/items"

# Make the API request
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()
    # Print the first few items in the dataset for inspection
    print(data)
else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")
```

```python
# Extract the relevant asset URL from the API response
items = data.get('features', [])

# Check if there are items in the response
if items:
    # Choose the first item (or any specific item you need)
    item = items[0]  # Adjust this if you have a specific item in mind
    asset_url = item['assets']['data']['href']  # This should be the URL to the data
    print(f"Asset URL: {asset_url}")
else:
    print("No items found in the dataset.")
```
