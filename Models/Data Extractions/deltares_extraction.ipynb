```python
import requests
from tqdm import tqdm
import planetary_computer as pc
import xarray as xr
import pandas as pd

items = data.get('features', [])

if items:
    item = items[0]
    
    asset_url = item['assets']['data']['href']
    print(f"Asset URL: {asset_url}")

    signed_url = pc.sign(asset_url)
    response = requests.get(signed_url, stream=True)

    if response.status_code == 200:
        total_size = int(response.headers.get('content-length', 0))
        file_path = 'deltares_reservoir_data.nc'
        
        with open(file_path, 'wb') as f:
            with tqdm(total=total_size, unit='B', unit_scale=True, desc="Downloading") as pbar:
                for chunk in response.iter_content(chunk_size=1024):
                    f.write(chunk)
                    pbar.update(len(chunk))  
                    pbar.set_postfix({"Downloaded": f"{pbar.n / total_size * 100:.2f}%"})
        
        print(f"File downloaded successfully! Saved at {file_path}")
        
        ds = xr.open_dataset(file_path)
        ds_filtered = ds.sel(time=slice('2011-01-01', '2015-12-31'))
        df_filtered = ds_filtered.to_dataframe()
        water_quality_df = pd.read_csv('/mnt/data/water_quality_training_dataset.csv')
        water_quality_df['Sample Date'] = pd.to_datetime(water_quality_df['Sample Date'])
        merged_df = pd.merge(water_quality_df, df_filtered, how='inner', on=['Latitude', 'Longitude'])
        merged_df = merged_df[merged_df['Sample Date'].between('2011-01-01', '2015-12-31')]

        merged_df.head()
    else:
        print(f"Failed to download the file. Status code: {response.status_code}")
else:
    print("No items found in the dataset.")
```

```python
import requests
from tqdm import tqdm
import planetary_computer as pc
import xarray as xr
import pandas as pd
import concurrent.futures

def download_year_data(year):
    items = data.get('features', [])
    
    if items:
        item = items[0]
        asset_url = item['assets']['data']['href'] 
        print(f"Asset URL for {year}: {asset_url}")
        
        signed_url = pc.sign(asset_url)
        
        try:
            response = requests.get(signed_url, stream=True)
            if response.status_code == 200:
                total_size = int(response.headers.get('content-length', 0))
                file_path = f'deltares_reservoir_data_{year}.nc'
                with open(file_path, 'wb') as f:
                    with tqdm(total=total_size, unit='B', unit_scale=True, desc=f"Downloading {year}") as pbar:
                        for chunk in response.iter_content(chunk_size=8192):
                            f.write(chunk)
                            pbar.update(len(chunk))  # Update the progress bar with the chunk size
                            pbar.set_postfix({"Downloaded": f"{pbar.n / total_size * 100:.2f}%"})
                
                print(f"File for {year} downloaded successfully! Saved at {file_path}")
                return file_path
            else:
                print(f"Failed to download data for {year}. Status code: {response.status_code}")
                return None
        except Exception as e:
            print(f"Error downloading data for {year}: {e}")
            return None
    else:
        print("No items found in the dataset.")
        return None


# Function to filter and merge the data for each year
def filter_and_merge(year):
    file_path = download_year_data(year)
    
    if file_path:
        ds = xr.open_dataset(file_path)
        ds_filtered = ds.sel(time=slice(f'{year}-01-01', f'{year}-12-31'))
        df_filtered = ds_filtered.to_dataframe()
        water_quality_df = pd.read_csv('/mnt/data/water_quality_training_dataset.csv')
        water_quality_df['Sample Date'] = pd.to_datetime(water_quality_df['Sample Date'])
        merged_df = pd.merge(water_quality_df, df_filtered, how='inner', on=['Latitude', 'Longitude'])
        merged_df = merged_df[merged_df['Sample Date'].between(f'{year}-01-01', f'{year}-12-31')]
        
        return merged_df
    else:
        print(f"Skipping year {year} due to download failure.")
        return None


def download_and_merge_all_years(years=[2011, 2012, 2013, 2014, 2015]):
    with concurrent.futures.ThreadPoolExecutor(max_workers=5) as executor:
        futures = {executor.submit(filter_and_merge, year): year for year in years}
        
        merged_data = []
        
        for future in concurrent.futures.as_completed(futures):
            year = futures[future]
            try:
                result = future.result()
                if result is not None:
                    merged_data.append(result)
            except Exception as e:
                print(f"Error processing data for {year}: {e}")
        
        if merged_data:
            all_data = pd.concat(merged_data)
            print(f"All data for {years} has been merged successfully.")
            return all_data
        else:
            print("No data was successfully downloaded and merged.")
            return None


final_merged_data = download_and_merge_all_years()

if final_merged_data is not None:
    final_merged_data.head()
```
